```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='NanumGothic')  # 한글 폰트 사용
plt.rcParams['axes.unicode_minus'] = False  # 음수 부호 깨짐 방지

# 그래프 시각화를 위한 함수 정의
def plot_function_relation(mapping, title):
    G = nx.DiGraph()  # 방향 그래프 생성

    # 노드 추가 (집합 X와 Y)
    for x in mapping.keys():
        G.add_node(f"X{x}", pos=(0, -x))  # X축에 위치
    for y in mapping.values():
        G.add_node(f"Y{y}", pos=(1, -ord(y) + ord('a')))  # Y축에 위치

    # 엣지 추가 (X -> Y 매핑)
    for x, y in mapping.items():
        G.add_edge(f"X{x}", f"Y{y}")

    # 그래프 그리기
    pos = nx.get_node_attributes(G, 'pos')
    nx.draw(
        G, pos, with_labels=True, node_color="lightblue",
        node_size=2000, font_size=10, arrows=True, edge_color="gray"
    )
    plt.title(title)
    plt.axis('off')

# (a) 함수
mapping_a = {1: 'a', 2: 'b', 3: 'c'}
plt.subplot(1, 3, 1)
plot_function_relation(mapping_a, "(a) 함수")

# (b) 함수
mapping_b = {1: 'a', 2: 'a', 3: 'c'}
plt.subplot(1, 3, 2)
plot_function_relation(mapping_b, "(b) 함수")

# (c) 함수
mapping_c = {1: 'a', 2: 'b', 3: 'a', 4: 'c'}
plt.subplot(1, 3, 3)
plot_function_relation(mapping_c, "(c) 함수")

plt.tight_layout()
plt.show()
```
